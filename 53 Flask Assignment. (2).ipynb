GET and POST methods:

The GET and POST methods are two of the most common HTTP request methods used for transferring data
between a client and server.

The GET method is used to request data from the server. When a client makes a GET request,
it retrieves data from the server without making any changes to the data on the server.
For example, when a user enters a URL into their browser, the browser makes a GET request
to the server to retrieve the web page associated with that URL.

The POST method is used to submit data to the server for processing. When a client makes
a POST request, it sends data to the server that can be used to create, update, or delete data on the server. For example, when a user submits a form on a website, the form data is typically sent to the server using a POST request.

In Flask, the request object is used to access incoming request data from the client. 

The request object contains information about the incoming request, such as the HTTP method,
headers, and data.

The request object can be used to access request data in different formats, such as form data,
JSON, or file uploads. The request object also provides methods to access query parameters, 
request headers, and cookies.

The redirect() function in Flask is used to redirect the user to a different URL. 
This is useful for implementing login/logout functionality, or for redirecting the user
to a different page after they have submitted a form.

To use the redirect() function, we can simply pass the URL that you want to redirect the user
to as an argument.

Templates in Flask are used to generate HTML pages dynamically.
A template is a file that contains HTML markup with placeholders for dynamic content.
When a request is made, Flask uses the template engine to replace these placeholders with
actual data, and then sends the resulting HTML to the client.

The render_template() function in Flask is used to render templates. It takes the name of
the template file as an argument, and returns the rendered HTML. For example:

from flask import Flask, render_template

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return render_template('hello.html', name=name)
